```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing import sequence
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, Dense

```

```python
## Mapping or words index back to words (for understanding)
word_index = imdb.get_word_index()
# reverse word_index
reverse_word_index = {value: key for key,value in word_index.items()}
```

```python
# load the pretrained model with ReLU activation function
from tensorflow.keras.models import load_model
model = load_model('simple_rnn_imdb_model.h5')
model.summary()
```

```python
# we can also get weights with respect to each layer or parameters of the model which is used for training the model and making predictions
model.get_weights()
```

```python
model.layers[1].get_weights()[1]
```

```python
## step2-> helper function


# function to decode the review back to words
def decode_review(text):
    return ' '.join([reverse_word_index.get(i-3,'?') for i in text])

# function to preprocess  the  user input review
def preprocess_text(text):
    # split the text into words
    words = text.lower().split()
    # convert the words to integers using the word_index
    integers = [word_index.get(word, 2)+ 3 for word in words]
    # pad the sequence to the same length as the training data
    padded_sequence = sequence.pad_sequences([integers], maxlen=500)
    return padded_sequence
```

```python
### prediction function

def predict_sentiment(review):
    # preprocess the text
    preprocessed_text = preprocess_text(review)
    # make a prediction using the model
    prediction = model.predict(preprocessed_text)
    # return the predicted sentiment
    return 'positive' if prediction[0][0] > 0.5 else 'negative', prediction[0][0]
```

```python
# step 4 user input and prediction
# example review for prediction
user_review = "This movie was fantastic! I really enjoyed it."
print(predict_sentiment(user_review))
```
