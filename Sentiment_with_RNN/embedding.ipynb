```python
# First we convert sentences into numbers because neural networks understand only numbers. Then the embedding layer converts each word index into a dense vector that captures the meaning of the word. The sequential model organizes neural network layers in a pipeline where data flows from one layer to the next until the final prediction is produced.
from tensorflow.keras.preprocessing.text import one_hot
```

```python
### sentences
sentence=[  'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good',]
```

```python
print(sentence)
```

```python
# define the vocabulary size
voc_size=10000
```

```python
### one hot representation of the sentences
one_hot_repr=[one_hot(words,voc_size) for words in sentence]
print(one_hot_repr)
```

```python
one_hot_repr[0]
```

```python
### word embedding representation
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
## define maximum length of the sentence
max_sent_length=8

embedding_docs = pad_sequences(one_hot_repr,maxlen=max_sent_length,padding='pre')
```

```python
print(embedding_docs)
```

```python
# feature representation using embedding
dim = 10
```

```python
model = Sequential()
model.add(Embedding(voc_size,dim,input_length=max_sent_length))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
model.predict(embedding_docs)
```

```python
print(sentence[0])
print(embedding_docs[0])
print(model.predict(embedding_docs[0]))
```
